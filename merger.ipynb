```python
import pandas as pd
import numpy as np
import os
import glob

data= 'outputs/2024-03-27/20-59-09'
```

```python
print(os.getcwd())
```

```python
dfs = []

for file in glob.glob(data + '/submission_test_*.csv'):
    print(file)
    df = pd.read_csv(file)
    print(df.head())
    dfs.append(df)
    # print(df.columns)
    # print(df.shape)
    # print(df.isnull().sum())
    # print(df.dtypes)
    # print(df.describe())
    # print(df.info())
    # print(df['id'].nunique())
    # print(df['id'].value_counts())
    # print(df['id'].unique)
```

```python
final_df = pd.DataFrame(columns=['Id','Predicted'])
for i in range(len(dfs[0])):
    zero = 0
    one = 0
    for j in range(len(dfs)):
        if dfs[j].iloc[i,1] == 0:
            zero += 1
        else:
            one += 1
    print(dfs[0].iloc[i,0], zero, one)
    if zero > one:
        final_df.loc[len(final_df)] = [dfs[0].iloc[i,0],0]
        # final_df = final_df.append({'Id':dfs[0].iloc[i,0],'Predicted':0},ignore_index=True)
    else:
        final_df.loc[len(final_df)] = [dfs[0].iloc[i,0],1]
        # final_df = final_df.append({'Id':dfs[0].iloc[i,0],'Predicted':1},ignore_index=True)
```
